```python
# used in the simulated contamination

contam = {}
contam_list = open("results/agg_contam/all_contamination_stats.txt")
contam_list.readline()

for line in contam_list:
    sp = line.strip().split()
    contam[int(sp[0])] = int(sp[1])

contiguity_identity = {}
assembly_stats = open("/analysis/2021_08_26_PlasmidSeq_paper/scripts/contamination/results/plasmid_stat/all_rotated.stats")
assembly_stats.readline()
for line in assembly_stats:
    sp = line.strip().split()
    # print(str(sp[4]) + " + " + str(sp[4]))
    contiguity_identity[int(sp[0].split("_")[0])] = [float(sp[4]),float(sp[5])]
    
contam_orig = {}
contam_list_orig = open("results/agg_contam_orig/all_contamination_stats.txt")
contam_list_orig.readline()

for line in contam_list_orig:
    sp = line.strip().split()
    contam_orig[int(sp[0])] = int(sp[1])

input_cont = open("computational_96_well_mixture_addgene.txt")
output_cont = open("computational_96_well_mixture_addgene_with_contam_2021_12_06_fragment_adj.txt","w")
output_cont.write(input_cont.readline().strip() + "\tcontamination\tcontamination_orig\tcontam_set_rate\trep\tplasmid\tcontiguity\tidentity\n")
for line in input_cont:
    sp = line.strip().split("\t")
    contam_rate = str(float(sp[1].split("_")[3].replace("p",".")) * 100.0)
    est = float(contam[int(sp[0])])
    est_orig = float(contam_orig[int(sp[0])])
    contiguity = contiguity_identity[int(sp[0])]
    plasmid = sp[1].split("_")[0]
    output_cont.write(line.strip() + "\t" + str(est) + "\t" + str(est_orig) + "\t" + contam_rate + "\t" + sp[1].split("_")[5] + "\t" + plasmid + "\t" + str(contiguity[0]) + "\t" + str(contiguity[1]) + "\n")

output_cont.close()
```
